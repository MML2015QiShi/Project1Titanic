###Deal with missing values
We can see that there are missing values for some of the varibles like Age, Cabin and Embarked. We can try to deal with them by filling the missing values.

To fill missing age values, we look at the group of people who has a missing age:

```python
df[df['Age'].isnull()][['Sex', 'Pclass', 'Age']]
```

Instead of just filling the median age for this whole group, we can split the group into sub-groups with different sex, passenger class and fill the median age for the sub-groups.

```python
#creating a new variable for Sex variable with numerical values:
df['Gender'] = df['Sex'].map( {'female': 0, 'male': 1} ).astype(int) 
```

```python
median_ages = np.zeros((2,3)) 
```

```python
for i in range(0, 2):
    for j in range(0, 3):
        median_ages[i,j] = df[(df['Gender'] == i) & (df['Pclass'] == j+1)]['Age'].dropna().median()
median_ages
```

```python
df['AgeFill'] = df['Age']
for i in range(0, 2):
    for j in range(0, 3):
        df.loc[ (df.Age.isnull()) & (df.Gender == i) & (df.Pclass == j+1), 'AgeFill'] = median_ages[i,j]
```

Next we fill the missing Embarked variables, since it only has 2 missing values, we just replace it instead of creating a new variable:

```python
df[df['Embarked'].isnull()]['Embarked'] = df['Embarked'].mode()
```

Now we only have cabin variable which has missing values. Since cabin numbers have too many missing values (687 out of 891), it might be better to just drop this variable for our analysis.

###Identify important factors
we want to find out the variables that has a clear impact on the survival, so we can use those variables to build a model to made predictions.

####Gender:

```python
temp3=pd.crosstab(df['Gender'],df['Survived']).apply(lambda r:r/r.sum().astype(float),axis=1)
temp3.plot(kind='bar',stacked=True)
```

It is clear that female passengers has a significant larger survival rate overall than male passengers.

####Age:

```python
barsize = 10
df['AgeGrouped']=np.floor(df['AgeFill']/barsize)*barsize
temp=pd.crosstab(df['AgeGrouped'],df['Survived']).apply(lambda r:r/r.sum().astype(float),axis=1)
temp.plot(kind='bar',stacked=True)
```

From the above graph, it looks like younger people may be more likely to survive comparing with older people.

####Pclass:
we now look at the Pclass variable, which has value 1,2,3 and represents the types of ticket class each passenger has. 

```python
temp0=pd.crosstab(df['Pclass'],df['Survived']).apply(lambda r:r/r.sum(),axis=1)
temp0.plot(kind='bar',stacked=True)
```

We can see a clear patter from the above graph: passenger in the better class has a higher survial rate.

####Embarked:
The Embarked variable indicate the physical location where the passenger got on board.

```python
temp1=pd.crosstab([df.Embarked,df.Pclass],df['Survived']).apply(lambda r:r/r.sum(),axis=1)
```

```python
temp1.plot(kind='bar',stacked=True)
```

From above grpah, there isn't a clear pattern of the relationship between varaible embarked and survival rate if we group by classes. So we'll just drop Embarked variable for the analysis now and come back later.

Now we look at the impact of Family size to the survival rate:

```python
df['FamilySize'] = df['SibSp'] + df['Parch'] #adding a family size variable
```

```python
temp2=pd.crosstab(df.FamilySize,df.Survived).apply(lambda r:r/r.sum(),axis=1)
```

```python
temp2.plot(kind='bar',stacked=True)
```

We can see that people with 1-3 family members are most likely to survive.
